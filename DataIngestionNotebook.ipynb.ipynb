#data transformation

```python
#  Filter Irrelevant Data
from pyspark.sql.types import StructType, StructField, IntegerType, StringType, FloatType, DateType
from pyspark.sql.functions import col, year

# Filter irrelevant data for accounts ( remove rows with zero or negative balances)
accounts_df_cleaned = accounts_df.filter(col("Balance") > 0)

# Filter irrelevant data for customers ( Remove rows where CustomerID is null or Name is invalid)
customers_df_cleaned = customers_df.filter((col("customer_id").isNotNull()) & (col("first_name").isNotNull()) & (col("last_name") != ""))

# Filter irrelevant data for loan payments (remove payments with non-positive amounts)
loan_payments_df_cleaned = loan_payments_df.filter(col("payment_amount") > 0)

# Filter irrelevant data for loans ( remove loans with non-positive principal amounts or without a valid loan ID)
loans_df_cleaned = loans_df.filter((col("loan_id").isNotNull()) & (col("loan_amount") > 0))

# Filter irrelevant data for transactions ( remove transactions with zero or negative amounts and invalid dates)
transactions_df_cleaned = transactions_df.filter((col("transaction_amount") > 0) & (col("transaction_date").isNotNull()))

```
